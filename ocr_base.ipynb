```python
from paddleocr import PaddleOCR

ocr = PaddleOCR(
    use_doc_orientation_classify=True, # 通过 use_doc_orientation_classify 参数指定不使用文档方向分类模型
    use_doc_unwarping=True, # 通过 use_doc_unwarping 参数指定不使用文本图像矫正模型
    use_textline_orientation=True, # 通过 use_textline_orientation 参数指定不使用文本行方向分类模型
)
```

```python
result = ocr.predict(r"e:\OneDrive\project_code\output\vision_temp\pdf_images_42020_5047\page_098.png")
# for res in result:
#     res.print()
#     res.save_to_img("output")
#     res.save_to_json("output")
```

位置识别

```python
import numpy as np
from pathlib import Path

data = result[0]                                   # 你的结果
boxes = np.array(data['dt_polys'])                 # 四点框
texts = data['rec_texts']

# 1. 计算每个框的中心 y，水平合并阈值 10 px
y_c = boxes.mean(axis=1)[:, 1]
rows = []
for i in np.argsort(y_c):                          # 从上到下
    if not rows or abs(y_c[i] - rows[-1]['y']) > 10:
        rows.append({'y': y_c[i], 'cells': []})
    rows[-1]['cells'].append((boxes[i, :, 0].min(), texts[i]))

# 2. 同一行按 x 排序后拼接
lines = []
for r in rows:
    line = ' '.join(t for _, t in sorted(r['cells'], key=lambda x: x[0]))
    lines.append(line)

clean = '\n'.join(lines)
print(clean)                                       # 直接看
# Path('page_116_clean.txt').write_text(clean, encoding='utf-8')  # 也落盘
```

```python
from paddleocr import PPStructure

# 版面识别 + OCR（支持中文）
table_engine = PPStructure(
    layout=True,           # 启用版面分析
    table=True,            # 启用表格识别
    ocr=True,              # 启用 OCR
    lang='ch'              # 支持中文
)

result = table_engine('demo.pdf')  # 支持图片/PDF

# 保存为 Excel、Word、HTML
from paddleocr.save_structure_res import save_structure_res
save_structure_res(result, save_folder='./output')
```
